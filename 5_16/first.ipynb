```python
import os
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split # 訓練データと評価データの仕分け
from sklearn.model_selection import KFold # クロスバリデーション
import lightgbm as lgb # 勾配ブースティング決定木法]

```

```python
df_train = pd.read_csv('train.tsv', sep='\t')
df_train.head()
```

```python
df_test = pd.read_csv('test.tsv', sep='\t')
df_test.head()
```

```python
train_x = df_train.drop(['id','bot'], axis=1)
train_y = df_train['bot']
test_x = df_test.drop(['id'], axis=1)
```

```python
from sklearn.linear_model import Perceptron
model = Perceptron()
model.fit(train_x, train_y)
```

```python
y_pred = model.predict(test_x)
```

```python
y_pred
```

```python
# submit.csvファイルの作成
df_submit = pd.DataFrame()
df_submit['id'] = df_test.iloc[:,[0]]
df_submit['result'] = y_pred
df_submit.head()
```

```python
df_submit.to_csv('submit.csv', header=False, index=False)
```
