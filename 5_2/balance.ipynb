```python
import pandas as pd
```

```python
df_train = pd.read_csv('data/train.tsv', index_col=0, sep='\t')
df_test = pd.read_csv('data/test.tsv', index_col=0, sep='\t')
```

```python
columns_l = list(df_train.columns)
y= columns_l.pop(0)
x = columns_l
```

```python
x_train = df_train[x] # 説明変数のカラム
y_train = df_train[y] # 目的変数カラム

x_test = df_test[x]
```

xgboost: treeの勾配ブースティングによる分類モデル．kaggleでよく利用される


```python
import numpy as np
import pandas as pd
import xgboost as xgb
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold

clf = xgb.XGBClassifier(num_class=3)

acc = []
cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=0)

for fold, (train_idx, val_idx) in enumerate(cv.split(x_train, y_train)):
    t_x = x_train.iloc[train_idx, :]
    val_x = x_train.iloc[val_idx, :]

    t_y = y_train.iloc[train_idx]
    val_y = y_train.iloc[val_idx]

    clf = xgb.XGBClassifier(num_class=3)
    clf.fit(t_x, t_y)

    pred = clf.predict(val_x)
    acc.append(accuracy_score(val_y, pred))

print("acc:", np.mean(acc))

test_preds = clf.predict(x_test)

submission_df = pd.DataFrame({
    "id": x_test.index,  
    "class": test_preds
})

submission_df.to_csv("submission.csv", index=False)
```
